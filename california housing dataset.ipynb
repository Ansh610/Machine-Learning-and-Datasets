```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
```

```python
# Load dataset
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['MedHouseValue'] = california.target  # Target variable

# Display first few rows
df.head()

```

```python
# Summary statistics
df.describe()
```

```python
# Histogram for each feature
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.show()

# Box Plot to check outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.show()

```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['MedInc'], df['HouseAge'], df['MedHouseValue'], c=df['MedHouseValue'], cmap='coolwarm')
ax.set_xlabel('Median Income')
ax.set_ylabel('House Age')
ax.set_zlabel('Median House Value')
plt.title("3D Scatter Plot of Housing Data")
plt.colorbar(scatter)
plt.show()
```
