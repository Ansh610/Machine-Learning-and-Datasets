```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Load an image (Upload an image first)
image_path ="/content/abc.webp"
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert to B/W (Thresholding)
_, bw = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

# Display Original, Grayscale, and B/W Images
cv2_imshow(image)  # Original
cv2_imshow(gray)   # Grayscale
cv2_imshow(bw)     # Black & White (Thresholded)
```

```python
# Function to plot histograms
def plot_histogram(img, title, color=None):
    plt.figure(figsize=(6, 4))
    if color:
        for i, col in enumerate(color):
            hist = cv2.calcHist([img], [i], None, [256], [0, 256])
            plt.plot(hist, color=col)
    else:
        hist = cv2.calcHist([img], [0], None, [256], [0, 256])
        plt.plot(hist, color="black")

    plt.title(title)
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.show()

# Plot histograms
plot_histogram(bw, "B/W Histogram", color=['black'])
plot_histogram(gray, "Grayscale Histogram")
plot_histogram(image, "RGB Histogram", color=['blue', 'green', 'red'])
```

```python
# Check if the image is normalized (pixel values between 0-1)
def is_normalized(img):
    return img.max() <= 1.0 and img.min() >= 0.0

# Check grayscale image
is_norm = is_normalized(gray)
print("Is the grayscale image normalized?", is_norm)
```

```python
# Normalize the image if it's not already normalized
if not is_norm:
    gray_normalized = gray / 255.0  # Normalize to [0,1]
    print("Image normalized successfully!")
else:
    gray_normalized = gray

# Check again
print("Is normalized now?", is_normalized(gray_normalized))
```

```python
# Apply Histogram Equalization
equalized = cv2.equalizeHist(gray)

# Display original and equalized images
cv2_imshow(equalized)

# Plot histograms
plot_histogram(equalized, "Equalized Histogram")
```
